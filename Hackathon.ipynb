```python
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
import bibtexparser
from datetime import datetime

class PublicationSummaryGenerator:
    def __init__(self, root):
        self.root = root
        self.root.title("Publication Summary Generator")
        self.root.geometry("500x300")

        self.create_widgets()

    def create_widgets(self):
        tk.Label(self.root, text="Choose Input File:").pack(pady=10)

        tk.Button(self.root, text="Upload Excel File", command=self.load_excel).pack(pady=5)
        tk.Button(self.root, text="Upload BibTeX File", command=self.load_bibtex).pack(pady=5)

        tk.Label(self.root, text="Enter Start Year:").pack(pady=5)
        self.start_year_entry = tk.Entry(self.root)
        self.start_year_entry.pack(pady=5)

        tk.Label(self.root, text="Enter End Year:").pack(pady=5)
        self.end_year_entry = tk.Entry(self.root)
        self.end_year_entry.pack(pady=5)

        tk.Button(self.root, text="Generate Summary", command=self.generate_summary).pack(pady=20)

    def load_excel(self):
        self.file_path = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx *.xls")])
        self.data_source = 'excel'
        messagebox.showinfo("Success", "Excel file loaded successfully!")

    def load_bibtex(self):
        self.file_path = filedialog.askopenfilename(filetypes=[("BibTeX files", "*.bib")])
        self.data_source = 'bibtex'
        messagebox.showinfo("Success", "BibTeX file loaded successfully!")

    def generate_summary(self):
        start_year = int(self.start_year_entry.get())
        end_year = int(self.end_year_entry.get())

        if self.data_source == 'excel':
            df = pd.read_excel(self.file_path)
        elif self.data_source == 'bibtex':
            with open(self.file_path) as bibtex_file:
                bib_database = bibtexparser.load(bibtex_file)
            df = pd.DataFrame(bib_database.entries)
        else:
            messagebox.showerror("Error", "No valid file loaded.")
            return

        df = df[df['year'].astype(int).between(start_year, end_year)]

        journal_df = df[df['ENTRYTYPE'] == 'article'] if 'ENTRYTYPE' in df else df[df['type'] == 'journal']
        conf_df = df[df['ENTRYTYPE'] == 'inproceedings'] if 'ENTRYTYPE' in df else df[df['type'] == 'conference']

        journal_df.to_excel("Journal_Publications_Summary.xlsx", index=False)
        conf_df.to_excel("Conference_Publications_Summary.xlsx", index=False)

        journal_df.to_csv("Journal_Publications_Summary.doc", sep='\t', index=False)
        conf_df.to_csv("Conference_Publications_Summary.doc", sep='\t', index=False)

        messagebox.showinfo("Success", "Publication summary generated!")

if __name__ == "__main__":
    root = tk.Tk()
    app = PublicationSummaryGenerator(root)
    root.mainloop()
```
